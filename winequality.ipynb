now we see the distribution of various features in the dataset.
Aslo calculating the appropriate measures of central tendency for this features

```python
plt.figure(figsize=(11,6))
sns.histplot(data = data  , x = 'fixed acidity', color = 'orange',
             edgecolor='linen',  alpha=0.5 , bins=5 )
plt.title(' Histogram of fixed acidity')
plt.xlabel('fixed acidity')
plt.ylabel('count')
plt.show()
```

# Observation ----


1.   Histogram is normally distributed
2.   Maximum count of values for Fixed acidity is in between  6 to 8 which is much higher than lower count .
Further investigation is needed .



```python
# calculating the mean of fixed acidity with round
round(data['fixed acidity'].mean(),2)
```

```python
# calculating the median  of fixed acidity
data['fixed acidity'].median()
```

```python
plt.figure(figsize=(11,6))
sns.histplot(data = data  , x = 'fixed acidity', color = 'orange',
             edgecolor='linen',  alpha=0.5 , bins=5 )
plt.title(' Histogram of fixed acidity')
plt.xlabel('fixed acidity')
plt.ylabel('count')
plt.vlines(data['fixed acidity'].mean(),ymin=0,ymax=4000,color='blue',label='mean')
plt.vlines(data['fixed acidity'].median(),ymin=0,ymax=4000,color='red',label='median')
plt.legend()
plt.show()
```

## Observation--
we can now see that mean and median are clear representative of data


---
beacause mean and median are very close to each other
.So by this we can choose either mean or median as a central tendency.





```python
plt.figure(figsize=(11,6))
sns.histplot(data = data  , x = 'volatile acidity', color = 'green',
             edgecolor='linen',  alpha=0.5 , bins=5 )
plt.title(' Histogram of volatile acidity')
plt.xlabel('volatile acidity')
plt.ylabel('count')
plt.show()
```

# Observation---
This histogram is not well disributed . it **skewed** a little towards right.


*   AS we seen an skwness therefore we can check the distribution using distplot function








```python
# plot displot using 'volatile acidity' feature
plt.figure(figsize=(12,6))
sns.distplot(data['volatile acidity'],color = 'purple')
plt.title(' Displot of volatile acidity')
plt.xlabel('volatile acidity')
plt.ylabel('Density')
plt.show()
```

# Observation ----


*   The above plot shown the normal ditribution becouse their is no skewnwss weather the right or left

The normal disribution is desrived by the mean and the standard deviation . The normal disribution is often reffered as to be a 'bell curve' becouse of it shape:


*   the median and mean is equal
*   it have only one mode
*   it is symmetric , meaning it decreases the same amount on the left and the right of the centre

we can calaculate skewness by using skew() function in python






```python
data['volatile acidity'].skew()
```

# obersvation ----
As we can see that output is greater than 1 which means the data is possitively skewed

```python
round(data['volatile acidity'].mean(),2)
```

```python
data['volatile acidity'].median()
```

```python
plt.figure(figsize=(11,6))
sns.histplot(data = data  , x = 'volatile acidity', color = 'green',
             edgecolor='linen',  alpha=0.5 , bins=5 )
plt.title(' Histogram of volatile acidity')
plt.xlabel('volatile acidity')
plt.ylabel('count')
plt.vlines(data['volatile acidity'].mean(),ymin=0,ymax=4000,color='blue',label='mean')
plt.vlines(data['volatile acidity'].median(),ymin=0,ymax=4000,color='red',label='median')
plt.legend()
plt.show()
```

# Oberservation
AS we can see that median is close to each other and the diffrenece between them is very small
we can safely choose the mean as the measure of central tendency here
